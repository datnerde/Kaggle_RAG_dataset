```r


library(tidyverse)
library(Matrix)
library(dplyr)
library(Amelia)
library(data.table)
library(plyr)
library(lubridate)
library(lightgbm)
library(gbm)

ProcessCalendarVariable <- function(x){
  if(x <=9){return(paste0('0',x))}
  else {return(as.character(x))}
}

start_time <- Sys.time()

list.files(path = "../input")
cat('Read weather data...\n')
weather_train <- fread("/kaggle/input/ashrae-energy-prediction/weather_train.csv",nrows = Inf, verbose = FALSE)
weather_test <- fread("/kaggle/input/ashrae-energy-prediction/weather_test.csv",nrows = Inf, verbose = FALSE)

cat('Impute weather data...\n')

dim <- dim(weather_train)[1]
data <- rbind(weather_train,weather_test)



                  
data$date <- NULL
head(data)

timestamp <- data$timestamp
data$timestamp <- NULL


d <- data.frame(matrix(0L,nrow= dim(data)[1],ncol = dim(data)[2]))
colnames(d) <- colnames(data)
n <- 15
weather_imputed <- invisible(amelia(data, m=n))
for (i in 1:n){
  d <- d + data.matrix(weather_imputed$imputations[[i]])
}
weather_imputed <- (1/n)*d

weather_imputed <- cbind(weather_imputed,timestamp)

weather_imputed$wind_direction <- NULL
weather_imputed$wind_speed <- NULL
weather_imputed$sea_level_pressure <- NULL
weather_imputed$precip_depth_1_hr <- NULL #this one 0.03 + in validation
weather_imputed$cloud_coverage <- NULL
weather_imputed$temp <- log((weather_imputed$dew_temperature-min(weather_imputed$dew_temperature)+1))*log((weather_imputed$air_temperature-min(weather_imputed$air_temperature)+1))


weather_train_imputed <- weather_imputed[1:dim,]
weather_test_imputed <- weather_imputed[-(1:dim),]


rm(weather_imputed,weather_train,weather_test,data,timestamp,d,dim,i,n)
invisible(gc(reset = TRUE))

end_time <- Sys.time()
end_time - start_time
tail(weather_train_imputed)
tail(weather_test_imputed)

cat('Train weather data...\n')
sapply(weather_test_imputed, function(x) sum(is.na(x)))
sapply(weather_train_imputed, function(x) sum(is.na(x)))
tail(weather_test_imputed)
tail(weather_train_imputed)
       
       
cat('Read building data...\n')

building_data <- fread("/kaggle/input/ashrae-energy-prediction/building_metadata.csv",nrows = Inf, verbose = FALSE)
building_data$primary_use <- as.numeric(factor(as.character(building_data$primary_use)))

d <- data.frame(matrix(0L,nrow= dim(building_data)[1],ncol = dim(building_data)[2]))
colnames(d) <- colnames(building_data)
n <- 100
cat('Impute building data...\n')
building_data_imputed <- amelia(building_data, m=n)

for (i in 1:n){
  d <- d + data.matrix(building_data_imputed$imputations[[i]])
}

building_data_imputed <- (1/n)*d
building_data_imputed$year_built <- 2018-round(building_data_imputed$year_built)

building_data_imputed$square_feet[building_data_imputed$square_feet<0] <- 283
building_data_imputed$floor_count[building_data_imputed$floor_count<0] <- 1


head(building_data_imputed)
rm(d,building_data,i,n)
invisible(gc())
       
cat('Add features...\n')

BuildData <- function(train = train , building = building, weather  = weather , nrows = nrows , skip = skip){

start_time <- Sys.time()

headers <- names(read.csv(train,nrows=1))
train <- fread(train,nrows = nrows,skip=skip,sep = ",", dec = ".",header=T, col.names=headers)
dim(train)
head(train)
d<- join(train, building,by="building_id", type = "left")
df<- join(d, weather,by=c("site_id","timestamp"),type = "left")

rm(d)
gc(reset = TRUE)
df$hour <- hour(df$timestamp)
df$day_of_week <- wday(df$timestamp)
df$week <- week(df$timestamp)
df$square_feet <- log(df$square_feet)
df$r_year <- (year(df$timestamp)-2016)*365+month(df$timestamp)*31+day(df$timestamp)

df$row_id <- NULL

end_time <- Sys.time()
end_time - start_time


return(df)
}
       
cat('Build traindata...\n')

train_imputed<- BuildData (train = '/kaggle/input/ashrae-energy-prediction/train.csv' , 
                           building = building_data_imputed, weather  = weather_train_imputed , nrows = Inf,skip = 0)

train_imputed<-train_imputed[!(building_id <= 104 & meter == 0 & timestamp <= "2016-05-20"),]
train_imputed<-train_imputed[!(site_id == 0 & timestamp <= "2016-02-29"),]
train_imputed<-train_imputed[!(site_id == 0 & primary_use ==5 & timestamp <= "2016-05-20"),]
train_imputed<-train_imputed[!(site_id == 0 & primary_use ==10 & timestamp <= "2016-05-20"),]
train_imputed<-train_imputed[!(building_id == 119 & timestamp <= "2016-08-01" & timestamp >= "2016-07-27"),]
train_imputed<-train_imputed[!(building_id == 281 & timestamp <= "2016-11-04" & timestamp >= "2016-10-25"),]
train_imputed<-train_imputed[!(building_id == 681 & timestamp <= "2016-04-25" ),]
train_imputed<-train_imputed[!(building_id == 681 & timestamp == "2016-07-13" ),]
train_imputed<-train_imputed[!(building_id == 678 & timestamp <= "2016-07-09" & timestamp >= "2016-07-07"  ),]
train_imputed<-train_imputed[!(building_id == 783 & timestamp <= "2016-12-09"),]
train_imputed<-train_imputed[!(building_id == 752 & timestamp <= "2016-10-09" & timestamp >= "2016-10-05"),]
train_imputed<-train_imputed[!(building_id == 751 & timestamp <= "2016-02-02"),]
train_imputed<-train_imputed[!(building_id == 763 & timestamp <= "2016-10-09" & timestamp >= "2016-10-05"),]
train_imputed<-train_imputed[!(building_id == 789 & timestamp <= "2016-09-04" & timestamp >= "2016-09-03"),]
train_imputed<-train_imputed[!(building_id == 790 & timestamp <= "2016-09-04" & timestamp >= "2016-09-03"),]
train_imputed<-train_imputed[!(building_id == 792 & timestamp <= "2016-11-12" & timestamp >= "2016-11-11"),]
train_imputed<-train_imputed[!(building_id == 954 & timestamp <= "2016-08-07" ),]
train_imputed<-train_imputed[!(building_id == 987 & timestamp <= "2016-11-16" & timestamp >= "2016-11-15"),]
train_imputed<-train_imputed[!(building_id == 981 & timestamp <= "2016-11-16" & timestamp >= "2016-11-15"),]
 train_imputed<-train_imputed[!(building_id == 1099 & timestamp <= "2016-06-16"),]
train_imputed<-train_imputed[!(building_id == 1099 & timestamp <= "2016-11-12" & timestamp >= "2016-11-08"),]      


cats <- c("building_id", "site_id", "meter", "primary_use")



train_imputed$timestamp <- NULL
N <- nrow(train_imputed)
y <- log1p(as.numeric(train_imputed$meter_reading))
train_imputed[, meter_reading := NULL]
train_imputed <- data.matrix(train_imputed)
tail(train_imputed)

tsf <- caret::createTimeSlices(1:N,initialWindow = FALSE,(N-1)/4,skip=(N-1)/4-1)
seq_along(tsf)
length(y)

cat('Train ...\n')
models <- list()
imp <- data.table()

p <- list(boosting = "gbdt",
          objective = "regression_l2",
          metric = "rmse",
          nthread = 5,
          learning_rate = 0.1,
          num_leaves = 440,
          min_child_weight = 9,
          colsample_bytree = 0.35,
          lambda = 3)
       
for (i in 1:4){
  cat("\nFold:", i, "\n")
  idx <- tsf[[2]][[i]]

  xtrain <- lgb.Dataset(train_imputed[-idx, ], label = y[-idx],categorical_feature = cats)
  xval <- lgb.Dataset(train_imputed[idx, ], label = y[idx],categorical_feature = cats)
  dim(train_imputed[-idx, ])
  dim(train_imputed[idx, ])
  models[[i]] <- lgb.train(p, xtrain, 12000, list(val = xval), eval_freq = 100, early_stopping_rounds = 100)
  
  imp <- rbind(imp, lgb.importance(models[[i]]))
  
  rm(xtrain, xval)
  invisible(gc())
}


lgb.plot.importance(imp[, lapply(.SD, mean), by=Feature], ncol(train_imputed), cex=1)
invisible(gc())
rm(train_imputed,weather_train_imputed,y,i,idx,imp)
gc(reset = TRUE)
ls()
       
te<- BuildData (train = '/kaggle/input/ashrae-energy-prediction/test.csv' , building = building_data_imputed, 
                          weather  = weather_test_imputed , nrows = 10000000, skip = 0)
te$timestamp <- NULL
te <- data.matrix(te)
tail(te)
dim(te)

cat("Making predictions...\n")
pred_te <- sapply(models, function(m_lgb) expm1(predict(m_lgb, te)))
invisible(gc())
cat('Read submit data...\n')
headers <- names(read.csv("/kaggle/input/ashrae-energy-prediction/sample_submission.csv",nrows=1))
sub <-  fread("/kaggle/input/ashrae-energy-prediction/sample_submission.csv",nrows = 10000000, skip = 0,sep = ",", dec = ".",header=T, col.names=headers)
sub[, meter_reading := round(rowMeans(pred_te), 2)
    ][meter_reading < 0, meter_reading := 0]

fwrite(sub, "submission_1.csv")
rm(headers,sub,te,pred_te)
gc(reset = TRUE)
                  
te<- BuildData (train = '/kaggle/input/ashrae-energy-prediction/test.csv' , building = building_data_imputed, 
                          weather  = weather_test_imputed , nrows = 10000000, skip = 10000000)
te$timestamp <- NULL
te <- data.matrix(te)
tail(te)
dim(te)

cat("Making predictions...\n")
pred_te <- sapply(models, function(m_lgb) expm1(predict(m_lgb, te)))
invisible(gc())
cat('Read submit data...\n')
headers <- names(read.csv("/kaggle/input/ashrae-energy-prediction/sample_submission.csv",nrows=1))
sub <-  fread("/kaggle/input/ashrae-energy-prediction/sample_submission.csv",nrows = 10000000, skip =  10000000,sep = ",", dec = ".",header=T, col.names=headers)
sub[, meter_reading := round(rowMeans(pred_te), 2)
    ][meter_reading < 0, meter_reading := 0]

fwrite(sub, "submission_2.csv")
rm(headers,sub,te,pred_te)
gc(reset = TRUE)
                  
                  
te<- BuildData (train = '/kaggle/input/ashrae-energy-prediction/test.csv' , building = building_data_imputed, 
                          weather  = weather_test_imputed , nrows = 10000000, skip = 20000000)
te$timestamp <- NULL
te <- data.matrix(te)
tail(te)
dim(te)

cat("Making predictions...\n")
pred_te <- sapply(models, function(m_lgb) expm1(predict(m_lgb, te)))
invisible(gc())
cat('Read submit data...\n')
headers <- names(read.csv("/kaggle/input/ashrae-energy-prediction/sample_submission.csv",nrows=1))
sub <-  fread("/kaggle/input/ashrae-energy-prediction/sample_submission.csv",nrows = 10000000, skip = 20000000,sep = ",", dec = ".",header=T, col.names=headers)
sub[, meter_reading := round(rowMeans(pred_te), 2)
    ][meter_reading < 0, meter_reading := 0]

fwrite(sub, "submission_3.csv")
rm(headers,sub,te,pred_te)
gc(reset = TRUE)
                  
te<- BuildData (train = '/kaggle/input/ashrae-energy-prediction/test.csv' , building = building_data_imputed, 
                          weather  = weather_test_imputed ,nrows = Inf, skip = 30000000)
te$timestamp <- NULL
te <- data.matrix(te)
tail(te)
dim(te)

cat("Making predictions...\n")
pred_te <- sapply(models, function(m_lgb) expm1(predict(m_lgb, te)))
invisible(gc())
cat('Read submit data...\n')
headers <- names(read.csv("/kaggle/input/ashrae-energy-prediction/sample_submission.csv",nrows=1))
sub <-  fread("/kaggle/input/ashrae-energy-prediction/sample_submission.csv",nrows = Inf, skip = 30000000,sep = ",", dec = ".",header=T, col.names=headers)
sub[, meter_reading := round(rowMeans(pred_te), 2)
    ][meter_reading < 0, meter_reading := 0]

fwrite(sub, "submission_4.csv")
                  
                  
sub1 <-  fread("/kaggle/working/submission_1.csv")
sub2 <-  fread("/kaggle/working/submission_2.csv")
sub3 <-  fread("/kaggle/working/submission_3.csv")
sub4 <-  fread("/kaggle/working/submission_4.csv")
head(sub1)
sub <- rbind(sub1,sub2,sub3,sub4)
dim(sub)
file.remove("/kaggle/working/submission_1.csv")
file.remove("/kaggle/working/submission_2.csv")
file.remove("/kaggle/working/submission_3.csv")
file.remove("/kaggle/working/submission_4.csv")


fwrite(sub, "submission_m2.csv")
```
